```python
#2️⃣String Operations & Text Processing
```

```python
import pandas as pd
import re
import nltk
nltk.download('punkt')

#cleaning the data and removing the unnecessary html tags using clean_text function

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Function to tokenize and lowercase text
def tokenize_text(text):
    tokens = nltk.word_tokenize(text)
    return [token.lower() for token in tokens]

df['JobRequirment_cleaned'] = df['JobRequirment'].apply(clean_text)
df['JobRequirment_tokens'] = df['JobRequirment_cleaned'].apply(tokenize_text)
df['JobDescription_cleaned'] = df['JobDescription'].apply(clean_text)
df['JobDescription_tokens'] = df['JobDescription_cleaned'].apply(tokenize_text)
```

```python
#Date operations
```

```python
# Convert using automatic detection (mixed formats)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['OpeningDate'] = pd.to_datetime(df['OpeningDate'], errors='coerce')
df['Deadline'] = pd.to_datetime(df['Deadline'], errors='coerce')

# Display results
df.dtypes
```

```python
#Calculate:
#o Days left to apply → Deadline - OpeningDate.
#o Job start delay → StartDate - OpeningDate.
#o Extract year & month from dates (Year, Month)
```

```python
df['OpeningDate'] = pd.to_datetime(df['OpeningDate'], errors='coerce')
df['Deadline'] = pd.to_datetime(df['Deadline'], errors='coerce')
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')

# Ensure subtraction only happens for valid date values
df['DaysLeftToApply'] = (df['Deadline'] - df['OpeningDate']).dt.days

# Handle cases where StartDate is NaT (avoid TypeError)
df['JobStartDelay'] = (df['StartDate'] - df['OpeningDate']).dt.days.where(df['StartDate'].notna(), None)

# Extract year and month
df['Year'] = df['OpeningDate'].dt.year
df['Month'] = df['OpeningDate'].dt.month

# Display the result
df[['OpeningDate', 'Deadline', 'DaysLeftToApply', 'StartDate', 'JobStartDelay', 'Year', 'Month']]
```

```python
#4️⃣Splitting & Extracting Data
```

```python
#Split Location into City & Country.
```

```python
# Ensure Location column is a string
df['Location'] = df['Location'].astype(str)

# Split Location safely and handle missing values
df[['City', 'Country']] = df['Location'].str.split(' , ', n=1, expand=True)

# Fill missing values with "Unknown" (optional)
df['Country'] = df['Country'].fillna('Unknown')

# Display updated DataFrame
df[['Location', 'City', 'Country']]
```

```python
#Extract numeric salary range from Salary (e.g., "50K-60K USD" → min_salary=50000, max_salary=60000).
def extract_salary(salary_str):
    min_salary, max_salary = None, None  # Initialize variables

    match = re.findall(r'(\d+)K', salary_str)

    if len(match) == 2:
        min_salary, max_salary = int(match[0]) * 1000, int(match[1]) * 1000
    elif len(match) == 1:  # If only one salary value exists (e.g., "100K")
        min_salary, max_salary = int(match[0]) * 1000, int(match[0]) * 1000

    # Return None if no valid salary is found
    if min_salary is None or max_salary is None:
        return pd.Series([None, None])

    return pd.Series([min_salary, max_salary])

# Apply the extraction function
df[['Min_Salary', 'Max_Salary']] = df['Salary'].apply(lambda x: pd.Series(extract_salary(x)))

# Filter rows where both Min_Salary and Max_Salary are not None
filtered_df = df.dropna(subset=['Min_Salary', 'Max_Salary'])

filtered_df[['Salary', 'Min_Salary', 'Max_Salary']]
```

```python
#5️⃣Filtering & Categorization
```

```python
it_jobs = df[df['IT'] == True]

# Categorize Eligibility into Education Level
it_jobs['Education_Level'] = it_jobs['Eligibility'].apply(lambda x:
    'Bachelor’s' if 'Bachelor' in x else
    'Master’s' if 'Master' in x else
    'Other')

# Categorize Experience Range based on Job Requirements
it_jobs['Experience_Range'] = it_jobs['JobRequirment'].apply(lambda x:
    'Entry-level' if 'Entry' in str(x) else
    'Mid-level' if 'Mid' in str(x) else
    'Senior-level' if 'Senior' in str(x) else
    'Other')

# Display relevant columns
it_jobs[['Eligibility', 'Education_Level', 'JobRequirment', 'Experience_Range', 'Salary']]
```
